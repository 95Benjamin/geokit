```python
import geokit as gk
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create a RegionMask
rm = gk.RegionMask.load(   gk._test_data_["aachenShapefile.shp"]   )
rm
```

```python
# Find lines which cross into this region
geoms = rm.extractFeatures( gk._test_data_["aachen_rails.shp"] )
geoms.head()
```

```python
# Display these around the RegionMask
axh = rm.drawGeoms(geoms, figsize=(5,6), hideAxis=True, color='b')
plt.show()
```

```python
# Indicate RegionMask pixels within 500 meters of these lines
isNearToRail = rm.indicateFeatures( gk._test_data_["aachen_rails.shp"], buffer=500 )

# Visualize it!
axh = rm.drawImage(isNearToRail, figsize=(5,6), hideAxis=True, cmap='Reds', )
plt.show()
```
